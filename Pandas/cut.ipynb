# CUT  
Método usado para criar uma nova coluna com agrupamentos

```python
import pandas as pd
import numpy as np
#Para este estudo usaremos alguns dados do Titanic
df = pd.read_csv('titanic_train.csv')
df.drop(columns=['Ticket', 'Cabin'], inplace = True)
df.head()
```

### Dividir a coluna 'Age' em subgrupos classificadores  
## Criar uma nova coluna 

```python
df['faixa_etaria'] = pd.cut(df['Age'], bins = 4, labels= ('jovem', 'adulto', 'maduro', 'idoso'))
df.head()
```

```python
df.groupby('faixa_etaria')['Age'].agg(('count', 'median', 'min', 'max'))
```

```python
df.groupby(['faixa_etaria', 'Sex'])['Age'].agg(('count', 'median', 'min', 'max'))
```

```python
df.groupby(['faixa_etaria', 'Sex']).agg({'Survived': 'sum', 'Age':('min','max')})
```

```python
fet = df.groupby(['faixa_etaria', 'Sex']).agg({'Survived': 'sum', 'Age':('min','max')})
print(fet)
```

```python
fet_index = df.groupby(['faixa_etaria', 'Sex'], as_index=False).agg({'Survived': 'sum', 'Age':('min','max')})
print(fet_index)
```

```python
print(fet.shape) # Observar q tem 3 colunas
print(fet_index.shape)  # 5 colunas - o indice não é mais a tupla(faixa_etaria, sex)
```

> Note que em fet o indice é uma tupla (faixa_etaria, sex)
>>em fet_index o indice voltou a ser numerico, por isso faixa_etaria e sex voltaram a ser colunas
